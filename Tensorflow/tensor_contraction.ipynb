```python
def naive_relu(x):
  assert len(x.shape) == 2 # Tensor 2D인경우 바로 종료

  x - x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] = max(x[i, j], 0)

  return x
```

```python
def naive_add(x, y):
  assert len(x.shape) == 2 # Tensor 2D인경우 바로 종료
  assert x.shape == y.shape

  x = x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] += y[i, j]
  
  return x
```
